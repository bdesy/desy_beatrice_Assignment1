# a) Writing a routine to calculate the Bessel functions in integral form.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy import special

def bessel(n, x, precision=500):
    """
    :param: n is the order of the Bessel function desired
            x is the value where we want to evaluate the function
            precision is the number of little rectangles we want to add up to integrate
    :return: value of Bessel function of order n at point x """

    h = np.pi / precision
    thvals = np.arange(0, np.pi, h)
    result = 0
    for th in thvals:
        result += h * (np.cos(n*th - x * np.sin(th)))
    return result / np.pi
```

Now let's try it out by plotting the first five Bessel functions and compare them to the ones from the Scipy library "special".

```python
xvals = np.arange(0,20,0.01)
v = [1,2,3,4,5]
for vi in v:
    g = plt.plot(xvals, special.jn(vi, xvals), linestyle="--", label="homemade " + str(vi))
    j = plt.plot(xvals, bessel(vi, xvals), label="numpymade " + str(vi))
plt.legend(ncol=2)
plt.show()
```

At this scale, the dashed lines representing the homemade functions are precisely over the actual Bessel functions, oh happy day!

# b) Compute the Airy Disk

Let's suppose that the source is of wavelenght 500 nm, and that the telescope parameters are somehow normalized to R/2a = 1. In a similar state of mind, let's assume that the intensity of the point source is 1 at the center and 0 anywhere else.

```python
def airy_disk(q):
    """
    :param: q is the distance from the optical axis
    :return: value of intensity at a distance q from the axis"""
    
    l = 1
    I0 = 1
    x = np.pi * q / l
    return I0 * (2 * bessel(1,x) / x)**2

n = 200
dim = np.linspace(-4, 4, n)
X, Y = np.meshgrid(dim, dim)

Z = airy_disk(np.sqrt(X**2 + Y**2))
contraste = False
if contraste:
    for i in range(n):
        for j in range(n):
            if Z[i,j] > 1:
                Z[i,j] = 1

print Z.shape

plt.pcolormesh(X, Y, Z, cmap = "Greys")
plt.title("Airy disk, original contrast")
plt.show()
```

This seems to represent a point source indeed. Let's see if we can increase the contrast to discover interesting things.

```python
f = plt.figure(1)

Imaxs = [0.5, 0.2, 0.1, 0.01]
for I in Imaxs:
    nb = 221 + Imaxs.index(I)
    plt.subplot(nb)
    Z = airy_disk(np.sqrt(X**2 + Y**2))
    for i in range(n):
        for j in range(n):
            if Z[i,j] > I:
                Z[i,j] = I
    plt.pcolormesh(X, Y, Z, cmap = "Greys")
    plt.title("Airy disk, maximal intensity of " + str(I))
    plt.show()
```

We see that as we cut off the higher values of intensity, variations further away from the point source become more and more apparent. For now, the loss of information about high values, represented as the black maximal value on the plots, is well worth the study of the fine topology of the function.
