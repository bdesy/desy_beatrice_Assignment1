First step, let's see if we can compute Mandelbrot's set.

```python
import numpy as np, random as rd, matplotlib.pyplot as plt, matplotlib.cm as cm

def mandelbrot(c, n):
    """
    :param: c is a point in the complex plane
            n is the number of iteration
    :return: 0 or 1 depending if z converges or diverges"""
    
    z = 0
    for i in range(n):
        if (abs(z) < 2):
            z = z**2 + c
            if i == n-1:
                return 0
        else:
            return 1


dim = 200
truc = np.zeros((dim, dim))
for i in range(dim):
    for j in range(dim):
        x = -2 + (4. / dim)*i
        y = -2 + (4. / dim)*j
        c = x + y * 1j
        truc[i,j] = mandelbrot(c, 100)

plt.imshow(truc, cmap="viridis")
plt.title('Mandelbrot\'s set')
plt.show()
```

Turns out we totally can! Now let's add a little color to that.

```python
def mandelbrot2(c, n):
    """
    :param: c is a point in the complex plane
            n is the number of iteration
    :return: 0 or 1 depending if z converges or diverges"""
    
    z = 0
    res = 0
    for i in range(n):
        if (abs(z) < 2):
            z = z**2 + c
        else:
            res += 1
    return res

dim = 200
truc2 = np.zeros((dim, dim))
for i in range(dim):
    for j in range(dim):
        x = -2 + (4. / dim)*i
        y = -2 + (4. / dim)*j
        c = x + y * 1j
        truc2[i,j] = mandelbrot2(c, 100)

plt.imshow(truc2, cmap="viridis")
plt.show()
plt.title('mandelbrot, version 2')
```
