The derivative of $f(x) = x(x-1)$ is $$f'(x) = 2x-1$$ Numerically, we can write a derivative routine to compute this result at a specific point $x_0$.

```python
import numpy as np
import matplotlib.pyplot as plt

def deriv(func):
    """
    We need to define a function inside the derivative function in order to pass it arguments
    """
    def dfunc(x0, h=0.001):
        return (func((x0 + h)) - func(x0)) / h
    return dfunc

def f(x):
    return x * (x-1)

df = derivative(f)


h_s = np.logspace(-4, -14, num=11)
x = np.linspace(0, 2, 100)

plt.plot(x, f(x), "--", label = "f (x)")
for h in h_s:
    m = df(1,h)
    plt.plot(x, m * (x-1), label="h = " + str(h))
    plt.legend(ncol=2)

plt.show()
```

At this scale, all the derivatives computed seems to be at the same place, let's try zooming in to the point (1,0) to see if a noticable difference emerges.

```python
h_s = np.logspace(-4, -14, num=11)
x = np.linspace(0.9, 1.1, 100)

plt.plot(x, f(x), "--", label = "f (x)")
for h in h_s:
    m = df(1,h)
    plt.plot(x, m * (x-1), label="h = " + str(h))
    plt.legend(ncol=2)

plt.show()
```

And even if we get even closer,

```python
h_s = np.logspace(-4, -14, num=11)
x = np.linspace(0.999, 1.001, 100)

plt.plot(x, f(x), "--", label = "f (x)")
for h in h_s:
    m = df(1,h)
    plt.plot(x, m * (x-1), label="h = " + str(h))
    plt.legend(ncol=2)

plt.show()
```

the only noticable difference is that the original function also merges with all its derivatives.
