First, let's create all the classes and methods that we need.

```python
import random

class Ship:
    
    def __init__(self, name):
        self.name = "Ship " + name
        self.hullstr = 100
        self.shieldstr = 100
        self.laserpwr = 60
        self.destroyed = False
        self.deathly_turn = 0

    def is_destroyed(self):
        self.destroyed = True


    def __str__(self):
        if self.destroyed:
            return self.name + " has been destroyed."
        else:
            return self.name + " has a hull strengh of " + str(self.hullstr)\
                + " and a shield strenght of " + str(self.shieldstr)
    
    def is_shot(self, laserpwr):
        if self.shieldstr > 0:
            self.shieldstr -= laserpwr
            if self.shieldstr < 0:
                diff = 0 - self.shieldstr
                self.shieldstr = 0
                self.hullstr -= diff/2
        else :
            if self.hullstr > laserpwr / 2:
                self.hullstr -= laserpwr / 2
            elif self.hullstr <= laserpwr / 2:
                self.hullstr = 0
                self.is_destroyed()
        print(" - Now, ", self)
    
    def shoots(self, target):
        print(self.name + " shoots " + target.name)
        target.is_shot(self.laserpwr)

class Warship(Ship):
    
    def __init__(self, name):
        Ship.__init__(self, name)
        self.name = "Warship " + name
    
    def shoots(self, target):
        prob = random.random()
        if prob < 0.3:
            print(self.name + " shoots " + target.name + " with hih powered missile.")
            target.is_shot(self.laserpwr * 1.8)
        else:
            print(self.name + " shoots " + target.name)
            target.is_shot(self.laserpwr)
    
class Speeder(Ship):
    
    def __init__(self, name):
        Ship.__init__(self, name)
        self.name = "Speeder " + name
    
    def is_shot(self, laserpwr):
        prob = random.random()
        if prob < 0.5:
            print(" - " + self.name + " has dodged the shot!")
        else:
            if self.shieldstr > 0:
                self.shieldstr -= laserpwr
                if self.shieldstr < 0:
                    diff = 0 - self.shieldstr
                    self.shieldstr = 0
                    self.hullstr -= diff/2             
            else :
                if self.hullstr > laserpwr:
                    self.hullstr -= laserpwr / 2
                elif self.hullstr <= laserpwr / 2:
                    self.hullstr = 0
                    self.is_destroyed()
            print(" - Now,", self)
```
